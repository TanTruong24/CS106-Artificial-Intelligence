```python
import pandas as pd
import numpy as np
```

```python
name_stocks = ['AAPL', 'AXP', 'BA', 'CAT', 'CSCO', 'CVX', 'DD', 'DIS', 'GE', 'GS', 'HD', 'IBM', 'INTC', 'JNJ', 'JPM', 'KO', 'MCD', 'MMM', 'MRK', 'MSFT', 'NKE', 'PFE', 'PG', 'RTX', 'TRV', 'UNH', 'V', 'VZ', 'WMT', 'XOM']
```

```python
df = pd.read_csv('/content/gdrive/MyDrive/cs106_source/data/dow30_feb2018/{}.csv'.format('AAPL'))
df = df.drop(['Close'], axis=1)
df.rename(columns={'Adj Close': 'Close'}, inplace=True)
df['Tic'] = 'AAPL'

for i in range(1, len(name_stocks)):
  temp = pd.read_csv('/content/gdrive/MyDrive/cs106_source/data/dow30_feb2018/{}.csv'.format(name_stocks[i]))
  temp['Tic'] = name_stocks[i]
  temp = temp.drop(['Close'], axis=1)
  temp.rename(columns={'Adj Close': 'Close'}, inplace=True)
  print(temp.shape)

  df = pd.concat([df, temp])
```

```python
df.rename(columns={'Close': 'close', 'Date': 'date', 'Open': 'open', 'High': 'high', 'Low': 'low', 'Volume': 'volume', 'Tic': 'tic'}, inplace=True)
```
